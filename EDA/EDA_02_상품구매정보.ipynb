```python
p2_seasons = pd.read_csv('p2_seasons.csv', sep=',')
p2_seasons = p2_seasons.drop('Unnamed: 0', axis=1)
p2_seasons
```

```python
df4 = pd.read_csv('LPOINT_BIG_COMP_04_PD_CLAC.csv', sep=',')
```

```python
pdde_seasons = pd.merge(p2_seasons, df4, on='pd_c', how='inner')
pdde_seasons
```

```python
pdde_seasons = pdde_seasons.drop(['rct_no', 'cop_c', 'br_c', 'de_hr', 'buy_ct'], axis=1)
```

```python
pdde_seasons
```

```python
pdde_seasons.groupby('de_dt')['cust'].count()
```

```python
pdde_seasons_pivot = pdde_seasons.pivot_table(values='clac_hlv_nm', index='de_dt', columns='pd_nm', aggfunc='count', fill_value=0)
pdde_seasons_pivot = pdde_seasons_pivot.rename(index={1:'겨울', 2:'봄', 3:'여름', 4:'가을'})
pdde_seasons_pivot
```

```python
pdde_seasons_max = pd.DataFrame(pdde_seasons_pivot.idxmax())
pdde_seasons_max['seasons'] = pdde_seasons_max[0]
pdde_seasons_max = pdde_seasons_max.drop(0, axis=1)
pdde_seasons_max = pdde_seasons_max.reset_index()
pdde_seasons_max
```
