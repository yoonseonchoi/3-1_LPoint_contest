```python
import pandas as pd
import collections

df = pd.read_csv('cust_data.csv')
```

```python
df
```

```python
class cust_marketing:
    
    def __init__(self, data):
        self.df = data
        

    def cust_top3(self, train_cnt, train_pd):
        pd_list = []

        for i in range(len(train_cnt)):
            pd_list.extend([train_pd[i]]*int(train_cnt[i]))

        pd_cnt = collections.Counter(pd_list)
        train_top = pd_cnt.most_common(3)

        train_top3 = []
        for i in range(len(train_top)) :
            train_top3.append(train_top[i][0])

        return train_top3


    def pd_top5(self, df, category) :
        mcls = (df['clac_mcls_nm'] == category)
        df = df[mcls]
        pd = df['pd_nm'].values
        cnt = df['buy_ct'].values

        pd_list = []
        for i in range(len(pd)):
            pd_list.extend([pd[i]]*int(cnt[i]))

        pd_cnt = collections.Counter(pd_list)
        top5 = pd_cnt.most_common(5)

        for i in range(len(top5)) :
            print(" ", "{})".format(i+1), top5[i][0], end='')
        print()


    def cat_pd_top5(self, df, category):
        mcls = (df['clac_mcls_nm'] == category)
        df = df[mcls]
        obj = df['pd_nm'].values

        obj_cnt = collections.Counter(obj)
        top5 = obj_cnt.most_common(5)

        for i in range(len(top5)) :
            print(" ", "{})".format(i+1), top5[i][0], end='')
        print()


    def cat_top3(self, df, age, gender, grade):
        if 'buy_ct' in df.columns:
            data = df.drop(['buy_ct'], axis=1)
        else:
            data = df

        data = data.drop_duplicates()
        condition = (data['ma_fem_dv']==gender) & (data['ages']==age) & (data['grade']==grade)
        data = data[condition]
        pd = data['clac_mcls_nm'].values

        pd_cnt = collections.Counter(pd)
        top = pd_cnt.most_common(3)

        top3 = []
        for i in range(3) :
            top3.append(top[i][0])

        print("\n{}등급 {} {}이 가장 많이 찾은 상품".format(grade, age, gender))

        for i in range(len(top3)) :
            print("{}.".format(i+1), top3[i])
            self.cat_pd_top5(data, top3[i])


    def cust_target(self):
        cust_num = input("고객 번호를 입력하세요 : ")

        if cust_num not in self.df['cust'].values :
            print("\n신규 고객입니다.")
            age, gender = input("\n나이(20대, 30대, ..., 60대, 70대)와 성별(남성,여성)을 입력해 주세요.\n").split()
            grade = 'General'

            self.cat_top3(self.df, age, gender, grade)

        elif cust_num in self.df['cust'].values :
            data = self.df[self.df['cust']==cust_num]

            buy_cnt = data['buy_ct'].values
            pd = data['clac_mcls_nm'].values
            age = data['ages'].values[0]
            gender = data['ma_fem_dv'].values[0]
            grade = data['grade'].values[0]

            train_top = self.cust_top3(buy_cnt, pd)

            print("\n{}님을 위한 맞춤 추천 상품".format(cust_num))
            for i in range(len(train_top)) :
                print("{}. {}".format(i+1, train_top[i]))
                self.pd_top5(data, train_top[i])

            self.cat_top3(self.df, age, gender, grade)
```

```python
model = cust_marketing(df)
```

```python
import pickle

with open('cust_marketing.pickle', 'wb') as f:
    pickle.dump(model, f)
```
