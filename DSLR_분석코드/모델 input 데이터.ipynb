```python
import pandas as pd
import collections
import numpy as np

df_demo = pd.read_csv('LPOINT_BIG_COMP_01_DEMO.csv')
df_grade = pd.read_csv('cust_grade.csv')
df_pdde = pd.read_csv('LPOINT_BIG_COMP_02_PDDE.csv')
df_clac = pd.read_csv('LPOINT_BIG_COMP_04_PD_CLAC.csv')

df_1 = df_demo[['cust', 'ma_fem_dv', 'ages']]
cust_grade = df_grade[['cust', 'grade']]
df_cust = pd.merge(df_1, cust_grade, how='left', on='cust')
df_2 = df_pdde[['cust', 'pd_c', 'buy_ct']]
df_12 = pd.merge(df_cust, df_2, how='left', on='cust')
df_12 = df_12.dropna()
df_124 = pd.merge(df_12, df_clac, how='left', on='pd_c')
```
