```python
import pandas as pd
import collections
import numpy as np
```

```python
df_pdde = pd.read_csv('LPOINT_BIG_COMP_02_PDDE.csv')
df_clac = pd.read_csv('LPOINT_BIG_COMP_04_PD_CLAC.csv')
df_2 = df_pdde[['cust', 'pd_c', 'de_dt', 'buy_ct']]
df_24 = pd.merge(df_2, df_clac, how='left', on='pd_c')
```

```python
df_24
```

```python
def train_test(buy_cnt, pd):
    data_size = len(buy_cnt)
    train_size = round(data_size*0.8)
    
    train_cnt = buy_cnt[:train_size]
    test_cnt = buy_cnt[train_size:]
    train_pd = pd[:train_size]
    test_pd = pd[train_size:]
    
    return train_cnt, test_cnt, train_pd, test_pd


def train_top3(train_cnt, train_pd):
    pd_list = []

    for i in range(len(train_cnt)):
        pd_list.extend([train_pd[i]]*train_cnt[i])
            
    pd_cnt = collections.Counter(pd_list)
    train_top = pd_cnt.most_common(3)
    
    train_top3 = []
    for i in range(len(train_top)) :
        train_top3.append(train_top[i][0])
        
    return train_top3


def cust_acc(train, test):
    cnt = 0
    for i in train :
        if i in test :
            cnt += 1

    acc = round(cnt / len(train) * 100, 2)
    
    return acc


def total_acc(acc_list):
    acc_array = np.array(acc_list)
    acc_avg = np.mean(acc_array)
    
    return acc_avg
```

```python
cust_list = df_24.groupby('cust').size()[df_24.groupby('cust').size()>=88].index
```

```python
cust_list
```

```python
cust_list_1000 = np.random.choice(cust_list,1000, replace=False)

acc_list = []
df = df_24.sort_values('de_dt')

for i,j in enumerate(cust_list_1000) :
    data = df[df['cust']==j]
    
    buy_cnt = data['buy_ct'].values
    pd = data['pd_nm'].values
    
    train_cnt, test_cnt, train_pd, test_pd = train_test(buy_cnt, pd)
    
    train_top = train_top3(train_cnt, train_pd)
    
    acc = cust_acc(train_top, test_pd)
    acc_list.append(acc)
    
    if i%10 == 0:
        print(i,acc)
    
accuracy = total_acc(acc_list)
print(accuracy)
```
